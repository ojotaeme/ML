### 8. Remova as instâncias que possuem mais de um valor ausente (>=2)

```python
data_copy = data.copy()

for index, row in data.iterrows():
    null_count = row.isnull().sum()
    if null_count > 1:
        data_copy = data_copy.drop(index=index)

data_copy.shape
```

### 9. Para as instâncias que possuem apenas um valor ausente, faça imputação de dados de acordo com o atributo faltante (Categórico (Moda)/Numérico (Média)). Por que é necessário tratar valores ausentes?

```python
from pandas.api.types import is_numeric_dtype

for i in data_copy:
    if is_numeric_dtype(data_copy[i]):
        data_copy = data_copy.fillna(data_copy.mean(numeric_only=True))
    else:
        data_copy = data_copy.fillna(data_copy.mode().iloc[0])
```

### 10. Remova os atributos categóricos que contenham três ou mais valores distintos

```python
columns_to_drop = [col for col in data_copy.select_dtypes(include=['object']) if data_copy[col].nunique() >= 3]

data_copy = data_copy.drop(columns=columns_to_drop)

data_copy.shape
```

### 11. Transforme o conjunto em um novo conjunto via encoding (OneHotEncoding) (data_encoded). Qual a função do encoding?

Encoding é uma técnica usada para converter variáveis categóricas em uma forma numérica, o que é necessário para muitos algoritmos de aprendizado de máquina. Isso permite que os modelos entendam e trabalhem com variáveis categóricas de maneira eficaz.

```python
object_columns = data_copy.select_dtypes(['object']).columns

data_encoded = pd.get_dummies(data_copy, columns=object_columns)

print(data_encoded.head())
```

### 12. Aplique LOF e identifique os outliers (data_encoded). Quantos outliers existem? Remova os outliers. Porque, em alguns casos, é necessário remover outliers?

Em determinadas situações, a remoção de outliers é necessária pela sua influência nos cálculos estatísticos. Isso ocorre porque os valores atípicos são consideravelmente distintos do restante do conjunto de dados, podendo distorcer as análises e interpretações realizadas.

```python
from sklearn.neighbors import LocalOutlierFactor

LOF = None

def detect_outliers(data, n_neighbors=20, p=2):
    global LOF 
    LOF = LocalOutlierFactor(n_neighbors=n_neighbors, p=p)
    y_pred = LOF.fit_predict(data)
    return y_pred

# Usando a função para detectar outliers
y_pred = detect_outliers(data_encoded)
print(y_pred)
```

```python
X_scores = LOF.negative_outlier_factor_
X_scores
```

```python
LOFs_index = where(y_pred == -1)
LOFs_index
```

```python
outliers = data_encoded.iloc[LOFs_index]
outliers
```

```python
print(f"Temos {len(LOFs_index[0])} outliers")
```

```python
data_encoded = data_encoded.drop(outliers.index)
data_encoded
```

### 14. Aplique PCA (data_encoded) e mantenha apenas dois componentes. Por que é interessante aplicar técnicas como o PCA?

O PCA é uma técnica de redução de dimensionalidade que, como o nome sugere, diminui a quantidade de dados, simplificando assim a análise.

```python
dataframe = data_encoded.drop(columns=['CLASSE_neg', 'CLASSE_pos'])
print(dataframe)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
standardized_data = scaler.fit_transform(dataframe)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(standardized_data)

principal_components_vectors = pca.components_

principal_components_vectors
```

```python
pca.explained_variance_ratio_
```

```python
component_names = ['component {}'.format(i) for i in range(len(principal_components_vectors))]

components_df = pd.DataFrame(data=principal_components_vectors, index=component_names, columns=dataframe.columns)
print(components_df.head())
```

### 15. Faça o Scatter plot da distribuição das classes nas novas dimensões. O que vocês podem inferir a partir do plot gerado?

```python
import seaborn as sns

eixos = pca.fit_transform(standardized_data)

sns.scatterplot(x=eixos[:, 0], y=eixos[:, 1], hue=data_encoded['CLASSE_pos'])
```

Pode-se observar que as instâncias positivas e negativas exercem uma influência significativa no primeiro componente principal (PC1).
