## **Exercício 2**

Execute os algoritmos a seguir, com os respectivos hiperparâmetros, na busca pelos melhores hiperparâmetros, e avalie o melhor modelo em um conjunto de teste a parte (20% teste, 80% treinamento). Use 3-fold cv para seleção dos hiperparâmetros. Todo o processo será avaliado com a medida f1_weighted.

### **Carrega o dataset digits**

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits.data.shape
```

```python
#Divide digits em 20% teste e 80% treinamento
from sklearn.model_selection import train_test_split

digits_data = digits.data
digits_target = digits.target

dig_data_train, dig_data_test, dig_target_train, dig_target_test = train_test_split(digits_data, digits_target, test_size=0.2, random_state=0)

print("conjunto de treinamento", dig_data_train.data.shape)
print("conjunto de teste", dig_data_test.data.shape)
print("conjunto de resultados de treinamento", dig_target_train.data.shape)
print("conjunto de resultados de teste", dig_target_test.data.shape)
```

### **Carrega o dataset breast_cancer**

```python
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
breast_cancer.data.shape
```

```python
#Divide breast_cancer em 20% teste e 80% treinamento
from sklearn.model_selection import train_test_split

breast_cancer_data = breast_cancer.data
breast_cancer_target = breast_cancer.target

b_can_data_train, b_can_data_test, b_can_target_train, b_can_target_test = train_test_split(breast_cancer_data, breast_cancer_target, test_size=0.2, random_state=0)

print("conjunto de treinamento", b_can_data_train.data.shape)
print("conjunto de teste", b_can_data_test.data.shape)
print("conjunto de resultados de treinamento", b_can_target_train.data.shape)
print("conjunto de resultados de teste", b_can_target_test.data.shape)
```

### **Carrega o dataset wine**

```python
from sklearn.datasets import load_wine

wine = load_wine()
wine.data.shape
```

```python
from sklearn.model_selection import train_test_split

wine_data = wine.data
wine_target = wine.target

wine_data_train, wine_data_test, wine_target_train, wine_target_test = train_test_split(wine_data, wine_target, test_size=0.2, random_state=0)

print("conjunto de treinamento", wine_data_train.data.shape)
print("conjunto de teste", wine_data_test.data.shape)
print("conjunto de resultados de treinamento", wine_target_train.data.shape)
print("conjunto de resultados de teste", wine_target_test.data.shape)
```

```python
digits_melhores = []
breast_cancer_melhores = []
wine_melhores = []
```

## **DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

tree = DecisionTreeClassifier(random_state=0)
hiperparam_grid = {'criterion': ['gini','entropy'], 'max_depth': [3, 7, 9]}
```

#### Dataset: digits

```python
import pandas as pd

digits_best_tree = GridSearchCV(tree, hiperparam_grid, scoring=('f1_weighted'), cv=3) #Aqui tava com esse parâmetro antes: refit=True. Dps vejo se precisa disso ou não
dig_tree = digits_best_tree.fit(dig_data_train, dig_target_train)

df = pd.DataFrame(digits_best_tree.cv_results_)
df
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

print(digits_best_tree.best_params_)
print(digits_best_tree.best_score_)

dig_target_true, dig_target_pred = dig_target_test, dig_tree.predict(dig_data_test)
print(classification_report(dig_target_true, dig_target_pred))

teste = f1_score(dig_target_true, dig_target_pred, average='weighted')

best = ["tree", digits_best_tree.best_params_, teste]
digits_melhores.append(best)
print(digits_melhores)
```

#### Dataset: breast_cancer

```python
b_can_best_tree = GridSearchCV(tree, hiperparam_grid, scoring=('f1_weighted'), cv=3)
b_can_tree = b_can_best_tree.fit(b_can_data_train, b_can_target_train)

df = pd.DataFrame(b_can_best_tree.cv_results_)
df
```

```python
#Melhor configuração do DecisionTreeClassifier em breast_cancer

print(b_can_best_tree.best_params_)
print(b_can_best_tree.best_score_)

b_can_target_true, b_can_target_pred = b_can_target_test, b_can_tree.predict(b_can_data_test)
print(classification_report(b_can_target_true, b_can_target_pred))

teste = f1_score(b_can_target_true, b_can_target_pred, average='weighted')

best = ["tree", b_can_best_tree.best_params_, teste]
breast_cancer_melhores.append(best)
print(breast_cancer_melhores)
```

#### Dataset: wine

```python
wine_best_tree = GridSearchCV(tree, hiperparam_grid, scoring=('f1_weighted'), cv=3)
wine_tree = wine_best_tree.fit(wine_data_train, wine_target_train)

df = pd.DataFrame(wine_best_tree.cv_results_)
df
```
